```python
# load dataset
df = sns.load_dataset('tips')
# display first few
df.head()
```

```python
# draw a scatter plot between total_bill and tip
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
# create a regression line plot
sns.lmplot(x='total_bill', y='tip', data=df)    
```

```python
# select the feature column from df
X = df[['total_bill']]
# select the target column from df
y = df['tip']
```

```python
# load or call the linear regression model
model = LinearRegression()
# fit the model means train the model
model.fit(X, y)


```

```python
# predict tip for a new total_bill value
# for example, total_bill = 50
predicted_tip = model.predict([[10]])
print(f'Predicted tip for a total bill of $50: ${predicted_tip[0]}')
```

```python
# write the whole script above in one go
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
# load dataset
df = sns.load_dataset('tips')
# select feature and target
X = df[['total_bill']]
y = df['tip']
# create and train model
model = LinearRegression()
model.fit(X, y) 
# predict tip for a new total_bill value
predicted_tip = model.predict([[50]])
print(f'Predicted tip for a total bill of $50: ${predicted_tip[0]}')    
```
